```python
## Encode the encodable variables
## Q3
df['Q3'].replace('18-25',1,inplace=True)
df['Q3'].replace('26-35',2,inplace=True)
df['Q3'].replace('36-50',3,inplace=True)
df['Q3'].replace('50+',4,inplace=True)
```

```python
## Q5
df['Q5'].replace('No',0,inplace=True)
df['Q5'].replace('Yes(1)',1,inplace=True)
df['Q5'].replace('Yes(2-3)',2,inplace=True)
df['Q5'].replace('Yes(>3)',3,inplace=True)
```

```python
## Q34
df['Q34'].replace('No',0,inplace=True)
df['Q34'].replace('A little',1,inplace=True)
df['Q34'].replace('Yes',2,inplace=True)
```

```python
## Create the language subsets
finn_audio = df.iloc[:,22:29]
finn_video = df.iloc[:,50:57]

germ_audio = df.iloc[:,29:36]
germ_video = df.iloc[:,71:78]

greek_audio = df.iloc[:,36:43]
greek_video = df.iloc[:,57:64]

hind_audio = df.iloc[:,43:50]
hind_video = df.iloc[:,64:71]
```

```python
finn_audio = finn_audio.dropna()
finn_video = finn_video.dropna()
germ_audio = germ_audio.dropna()
germ_video = germ_video.dropna()
greek_audio = greek_audio.dropna()
greek_video = greek_video.dropna()
hind_audio = hind_audio.dropna()
hind_video = hind_video.dropna()
```
